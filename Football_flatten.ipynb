```python
import requests as re
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
from flatten_json import flatten
```

```python
class football_flatten:
    
    def __init__(self):
        
        self.page = None
        self.data = None
        self.n = None
        self.d1= None
        self.d2= None
        self.d3= None
   
    def load1(self,page):
        self.page=page
        f=open(f'page{page}.json')
        d=json.load(f)
        return d
        
    def process(data,n):
        
        #self.data=data
        #self.n=n
        data=data['response']
        y=[]
        for dictionary in data:
            y.append(flatten(dictionary))
        data1=json_normalize(y)
        data2=pd.DataFrame()
        data2=data1[['player_name','player_age','player_nationality',f'statistics_{n}_league_name',f'statistics_{n}_league_season',f'statistics_{n}_games_minutes',f'statistics_{n}_games_position',f'statistics_{n}_games_rating',f'statistics_{n}_goals_total',f'statistics_{n}_goals_assists',f'statistics_{n}_goals_saves',f'statistics_{n}_goals_conceded',f'statistics_{n}_passes_key',f'statistics_{n}_passes_accuracy',f'statistics_{n}_tackles_total',f'statistics_{n}_tackles_interceptions',f'statistics_{n}_duels_total',f'statistics_{n}_duels_won',f'statistics_{n}_dribbles_attempts',f'statistics_{n}_dribbles_success',f'statistics_{n}_penalty_scored',f'statistics_{n}_penalty_missed',f'statistics_{n}_penalty_saved']]
        data2=data2.rename(columns={'player_name':'Name','player_age':'Age','player_nationality':'Nationality',f'statistics_{n}_league_name':'League',f'statistics_{n}_league_season':'Season',f'statistics_{n}_games_minutes':'Minutes Played',f'statistics_{n}_games_position':'Position',f'statistics_{n}_games_rating':'Rating',f'statistics_{n}_goals_total':'Goals Scored',f'statistics_{n}_goals_assists':'Assists',f'statistics_{n}_goals_saves':'Goal Saves',f'statistics_{n}_goals_conceded':'Goals Conceded',f'statistics_{n}_passes_key':'Key Passes',f'statistics_{n}_passes_accuracy':'Pass Accuracy',f'statistics_{n}_tackles_total':'Total Tackles',f'statistics_{n}_tackles_interceptions':'Interceptions',f'statistics_{n}_duels_total':'Duels Attempted',f'statistics_{n}_duels_won':'Duels Won',f'statistics_{n}_dribbles_attempts':'Dribble Attempts',f'statistics_{n}_dribbles_success':'Successful Dribbles',f'statistics_{n}_penalty_scored':'Penalties Scored',f'statistics_{n}_penalty_missed':'Penalties Missed',f'statistics_{n}_penalty_saved':'Penalties Saved'})
        return data2
    
    def concat(d1,d2,d3):
        
        Data = pd.concat([d1,d2,d3], axis=0, join='inner',ignore_index=True)
        return Data
    
```

```python
p1=football_flatten()
data1=p1.load1(1)
```

```python
data2=p1.load1(2)
data3=p1.load1(3)
```

```python
d1=football_flatten.process(data1,0)
```

```python
d2=football_flatten.process(data2,0)
d3=football_flatten.process(data3,0)
```

```python
PL_data=football_flatten.concat(d1,d2,d3)
len(PL_data)
```

```python
d11=football_flatten.process(data1,1)
d12=football_flatten.process(data2,1)
d13=football_flatten.process(data3,1)
```

```python
League_cup=football_flatten.concat(d11,d12,d13)
League_cup.head()
```

```python
d21=football_flatten.process(data1,2)
d22=football_flatten.process(data2,2)
d23=football_flatten.process(data3,2)
```

```python
FAcup_data=football_flatten.concat(d21,d22,d23)
FAcup_data.head()
```

```python
d31=football_flatten.process(data1,3)
d32=football_flatten.process(data2,3)
d33=football_flatten.process(data3,3)
```

```python
UCL_data=football_flatten.concat(d31,d32,d33)
UCL_data.head()
```

```python
PL_data.to_csv('PL_MANU.csv')
League_cup.to_csv('Leaguecup_MU.csv')
FAcup_data.to_csv('FACUP_MU.csv')
UCL_data.to_csv('UCL_MU.csv')
```
